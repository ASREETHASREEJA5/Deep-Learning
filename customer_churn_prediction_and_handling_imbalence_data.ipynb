```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as pt
```

```python
df = pd.read_csv('/content/customer_churn.csv')
```

```python
df[:5]
```

```python
df.drop('customerID',axis = "columns",inplace = True)
```

```python
df[:5]
```

```python
df.dtypes
```

```python
df[pd.to_numeric(df.TotalCharges,errors = 'coerce').isnull()]
```

```python
df1 = df[df.TotalCharges !=" "]
```

```python
df1.TotalCharges = pd.to_numeric(df1.TotalCharges)
```

```python
df1.dtypes
```

```python
def print_unique_col_values(df):
  for column in df:
    if df[column].dtypes == 'object':
      print(f'{column}: {df[column].unique()}')
```

```python
print_unique_col_values(df1)
```

```python
df1.replace("No internet service",'No' ,inplace = True)
df1.replace("No phone service",'No' ,inplace = True)
```

```python
print_unique_col_values(df1)
```

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}')
```

```python
df1['gender'].replace({'Female':1, 'Male':0},inplace = True)
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}')
```

```python
df2 = pd.get_dummies(data=df1, columns=['InternetService','Contract','PaymentMethod'],dtype = int)
df2.columns
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

Train test split


```python
x = df2.drop('Churn',axis = 'columns')
y = df2['Churn']
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
x_tr[:5]
```

mode ANN using tensorflow

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_tr, y_tr, epochs=100)
```

```python
model.evaluate(x_te,y_te)
```

```python
y_pred = model.predict(x_te)
```

```python
y_pred[:5]
```

```python
y_p = []
for i in y_pred:
  if i >0.5:
    y_p.append(1)
  else:
    y_p.append(0)
```

```python
y_p[:5]
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_te,y_p))
```

model-1 under samplling

```python
count_class_0,count_class_1 = df1.Churn.value_counts()
df_class_0 = df2[df2['Churn'] == 0]
df_class_1 = df2[df2['Churn'] == 1]
```

```python
df_class_0_unders =df_class_0.sample(count_class_1)
df_class_0_train = pd.concat([df_class_0_unders,df_class_1],axis = 0)
print("under sampling")
print(df_class_0_train.Churn.value_counts())
```

```python
x = df_class_0_train.drop("Churn",axis = "columns")
y = df_class_0_train["Churn"]
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size = 0.2, random_state=15, stratify=y)
```

```python
x_tr.shape
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_tr, y_tr, epochs=100)
```

```python
model.evaluate(x_te,y_te)
```

```python
y_p = model.predict(x_te)
```

```python
y_pred = []
for i in y_p:
  if i>0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_te,y_pred))
```

2 over sampling

```python
df_class_1_over = df_class_1.sample(count_class_0,replace = True)
df_over = pd.concat([df_class_0,df_class_1_over],axis = 0)

print("over sampling")
print(df_over.Churn.value_counts())
```

```python
x = df_over.drop("Churn",axis = "columns")
y = df_over["Churn"]
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size = 0.2, random_state=15, stratify=y)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_tr, y_tr, epochs=100)
```

```python
model.evaluate(x_te,y_te)
y_p = model.predict(x_te)

y_pred = []
for i in y_p:
  if i>0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)

from sklearn.metrics import classification_report
print(classification_report(y_te,y_pred))
```
